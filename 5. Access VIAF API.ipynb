# Process
In this notebook, I will access the VIAF API.

## Import Libraries

```python
from viapy.api import ViafAPI, ViafEntity, SRUResult, SRUItem
import pandas as pd
import pickle as pkl
import time
```

## Read files

```python
books = pd.read_csv("data/items_books_for_viaf.csv", low_memory = False, index_col=0).fillna("")
```

```python
books["viaf_id"] = ""
```

```python
def get_viaf_entry(author_name):
    print(author_name)
    viaf = ViafAPI()
    viaf_id = ""
    potential_entries = viaf.suggest(author_name)[:5]
    if len(potential_entries)==0:
        print("No results!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!")
    for entry in potential_entries:
        if entry["nametype"]=="personal":
            print(entry["term"])
            try:
                viaf_id = entry["viafid"]
                print("We found viaf_id!")
            except:
                print(author_name + " has no Viaf ID.")
            break
        print("ISSUE! We didn't find for", author_name)
    return viaf_id
```

```python
definitive_unique_authors = books[books.correct_author!=""].drop_duplicates(subset="correct_author")
```

```python
definitive_unique_authors
```

```python
start = time.time()
i=0
for index, row in definitive_unique_authors.loc[252241:].iterrows():
    if row.viaf_id=="":
        author = row.correct_author
        viaf_id = get_viaf_entry(author)
        if viaf_id == "":
            print("try normal author name")
            viaf_id = get_viaf_entry(row.author)
        books.at[index, "viaf_id"] = viaf_id
        definitive_unique_authors.at[index, "viaf_id"] = viaf_id
    if i%500==0:
        print(index, time.time()-start)
        books.to_csv("data/items_books_after_viaf.csv")
    i+=1
```
