```python
from viapy.api import ViafAPI, ViafEntity, SRUResult, SRUItem
import pandas as pd
import pickle as pkl
import time
```

```python
books = pd.read_csv("data/items_books_after_viaf_and_filling_and_merging_and_more_viaf.csv", low_memory = False, index_col=0).fillna("")
```

```python
books_without_ISBN_results = books.loc[books.correct_author==""]
authors_with_viaf = books_without_ISBN_results[["author","alt_first_author","viaf_id"]].loc[books.viaf_id!=""]
```

```python
len(authors_with_viaf)
```

```python
authors_with_viaf
```

```python
books_without_ISBN_results = books_without_ISBN_results.drop("viaf_id", axis=1)
```

```python
books_without_ISBN_results_and_with_viaf = pd.merge(books_without_ISBN_results.reset_index(), authors_with_viaf, on=["author","alt_first_author"]).set_index("index")
```
